```python
from PIL import Image

def preprocess_image(image):
    desired_width = 64
    desired_height = 64
    resized_image = image.resize((desired_width, desired_height))
    normalized_image = np.array(resized_image) / 255.0
    return normalized_image
```

```python
import torch.nn as nn
import torch

class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1)
        self.relu = nn.ReLU()
        self.maxpool = nn.MaxPool2d(kernel_size=2, stride=2)
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1)

        output_shape = torch.Tensor(torch.zeros(1, 32, 16, 16)).size()
        output_size = output_shape[1] * output_shape[2] * output_shape[3]
        self.fc = nn.Linear(output_size, 2)
        
    def forward(self, x):
        x = self.conv1(x)
        x = self.relu(x)
        x = self.maxpool(x)
        x = self.conv2(x)
        x = self.relu(x)
        x = self.maxpool(x)
        x = x.view(x.size(0), -1)
        x = self.fc(x)
        return x
loss_fn = nn.CrossEntropyLoss()
model = CNN()    
```

```python
def classify_image(outputs):
    _, predicted = torch.max(outputs, 1)
    return predicted
```

```python
from torch.utils.data import DataLoader, Dataset
class YourTrainDataset(Dataset):
    def __init__(self, file_path, transform=None):
        self.file_path = file_path
        self.transform = transform

        with open(file_path, 'r') as f:
            lines = f.readlines()
            self.images = []
            self.labels = []
            for line in lines:
                image_name, class_id, species, breed_id = line.strip().split()
                self.images.append(image_name)
                self.labels.append(int(class_id))

    def __len__(self):
        return len(self.images)

    def __getitem__(self, index):
        image_name = self.images[index]
        label = self.labels[index]
        image_path = f'D:/project/oxford-iiit-pet/images/{image_name}.jpg'
        image = Image.open(image_path)

        if self.transform:
            image = self.transform(image)

        image = np.array(image)

        return image, label
```

```python
from torchvision.transforms import Compose, ToTensor, Resize
class YourTestDataset(Dataset):
    def __init__(self, file_path, transform=None):
        self.file_path = file_path
        self.transform = transform

        with open(file_path, 'r') as f:
            lines = f.readlines()
            self.images = []
            self.labels = []
            for line in lines:
                image_name, class_id, species, breed_id = line.strip().split()
                self.images.append(image_name)
                self.labels.append(int(class_id))

    def __len__(self):
        return len(self.images)

    def __getitem__(self, index):
        image_name = self.images[index]
        label = self.labels[index]
        image_path = f'D:/project/oxford-iiit-pet/images/{image_name}.jpg'
        image = Image.open(image_path)

        if self.transform:
            image = self.transform(image)

        image = np.array(image)

        return image, label
```
